```python
import fitz  # PyMuPDF
import pandas as pd
import re

# Open the PDF file
with fitz.open("1.pdf") as pdf:
    # Initialize an empty string to store extracted text
    text = ""
    
    # Extract text from pages 5 to 11 (index 4 to 10)
    for page_num in range(4, 11):
        page = pdf.load_page(page_num)
        text += page.get_text("text")

# Output or save the extracted text
print(text)

```

```python
# Split the text into lines
lines = text.split('\n')

# Remove the first 3 lines and the last line
lines = lines[6:-3]

# Join the remaining lines back into a single string
text = '\n'.join(lines)

# Output the modified text
print(text)
```

```python
parts = re.split(r'(?=\d+\.)', text)

# Remove any empty strings or unwanted entries
parts = [part.strip() for part in parts if part.strip()]

# Create a DataFrame from the extracted parts
df = pd.DataFrame(parts, columns=['Content'])

# Display the DataFrame
df
```

```python
def is_only_number(content):
    return content.strip().isdigit()

# Function to combine rows when a number is found and delete the row
i = 0
while i < len(df) - 1:  # Ensure we don't go out of range
    if is_only_number(df.loc[i, 'Content']):
        # Append the current row's content to the next row
        df.loc[i + 1, 'Content'] = df.loc[i + 1, 'Content'] + ' ' + df.loc[i, 'Content']
        # Drop the current row and reset the index
        df = df.drop(i).reset_index(drop=True)
        # After dropping the row, stay at the same index since the rows shift
    else:
        i += 1  # Move to the next row if no deletion was made

# Print the updated DataFrame
df
```

```python
df['Title'] = df['Content'].str.extract(r'\d+\.\s+([^\d]+)')
# Removing the title from the Content column
#df['Content'] = df['Content'].str.replace('\n', ' ', regex=False)
df['Content'] = df['Content'].str.replace(r'\d+\.\s+([^\d]+)', '', regex=True)
df
```

```python
# Split the 'Content' based on '\n' and expand into separate rows
df_exploded = df.set_index('Title')['Content'].str.split('\n', expand=True).stack().reset_index(level=1, drop=True)

# Create a new column 'sub_title' for the separated content
df_exploded = df_exploded.to_frame(name='sub_title')

# Reset index to keep 'Title' in each row
df_exploded['Title'] = df_exploded.index.get_level_values(0)

# Clean up the resulting DataFrame to ensure no 'Title' is in the 'sub_title' column
df= df_exploded[df_exploded['sub_title'] != 'Title'].reset_index(drop=True)
column_name = "sub_title"
if column_name not in df.columns:
    raise ValueError(f"Column '{column_name}' does not exist in the DataFrame!")

def is_sentence(row):
    return not row.strip().isdigit()

new_subtitles = []
skip_next = False

for i in range(len(df) - 1):
    if skip_next:
        skip_next = False
        continue

    current_row = df.loc[i, column_name]
    next_row = df.loc[i + 1, column_name]

    if is_sentence(current_row) and is_sentence(next_row):
        # Concatenate successive sentences
        new_subtitles.append(current_row + " " + next_row)
        skip_next = True  # Skip the next row since it's concatenated
    elif not skip_next:
        new_subtitles.append(current_row)

# Handle the last row if it was not concatenated
if not skip_next:
    new_subtitles.append(df.loc[len(df) - 1, column_name])

# Update the DataFrame
df[column_name] = new_subtitles + [""] * (len(df) - len(new_subtitles))  # Fill extra rows with empty strings


df
```

```python
import fitz  # PyMuPDF

# Open the PDF file
pdf_path = "1.pdf"  # Replace with your file path
doc = fitz.open(pdf_path)

# Define the RGB threshold for black
def is_black(color, threshold=200):
    """Check if a color is approximately black.
    Converts the color integer to (R, G, B).
    """
    # Convert color to RGB (it's stored as a single integer in 0xRRGGBB format)
    r = (color >> 16) & 0xFF  # Extract red component
    g = (color >> 8) & 0xFF   # Extract green component
    b = color & 0xFF          # Extract blue component
    
    # If the color values are close to black (all low RGB values), return True
    return r < threshold and g < threshold and b < threshold

# Initialize a list to store non-black text
non_black_text = []

# Specify the page range (12 to 19 corresponds to indices 11 to 18)
for page_num in range(11, 19):  # Page indices are 0-based
    page = doc[page_num]
    for block in page.get_text("dict")["blocks"]:  # Get blocks of text
        if "lines" in block:
            for line in block["lines"]:
                for span in line["spans"]:
                    # Check if the text color is not black
                    color = span.get("color")
                    if color and not is_black(color):  # Filter non-black text
                        non_black_text.append(span["text"])

# Combine and clean the extracted text
non_black_text = [text.strip() for text in non_black_text if text.strip()]  # Clean text

# Output the non-black text

print(non_black_text)

```

```python
# Convert non_black_text into individual words
non_black_words = set()  # Use a set to avoid duplicates
for text in non_black_text:
    non_black_words.update(text.split())  # Split each element into words and add them to the set

# Iterate through the rows of the 'sub_title' column
for index, row in df.iterrows():
    sub_title = row["sub_title"]
    # Iterate through each word in non_black_words
    for word in non_black_words:
        if word in sub_title:
            # Replace the word with an empty string if it exists in the 'sub_title'
            sub_title = sub_title.replace(word, "").strip()
    # Update the 'sub_title' column with the modified value
    df.at[index, "sub_title"] = sub_title
    
# Replace '\n' with a space in the 'Title' column
df["Title"] = df["Title"].str.replace("\n", " ", regex=False)

# Replace '&' with ' ' only at the start or end of the sentence in the 'sub_title' column
df['sub_title'] = df['sub_title'].apply(lambda x: re.sub(r'^\&|\&$', ' ', x) if isinstance(x, str) else x)

# Display the updated DataFrame to verify the changes



# Output the modified DataFrame
df
```

```python
import fitz  # PyMuPDF
import pandas as pd

# Function to extract text from the PDF
def extract_text_from_page(pdf_path, page_num):
    doc = fitz.open(pdf_path)
    page = doc.load_page(page_num)
    # Extract text from the page and replace \n with a space
    page_text = page.get_text("text")
    return page_text.replace('\n', ' ')  # Replace newline characters with spaces

# Function to extract content based on subtitle index or range of pages
def extract_content(pdf_path, start_page, subtitle_index, next_start_page, next_subtitle_index):
    doc = fitz.open(pdf_path)
    
    if subtitle_index is not None:  # Case 1: Extract between subtitle indices
        # Extract text from start_page
        start_page_text = extract_text_from_page(pdf_path, start_page)
        start_content = start_page_text[subtitle_index:]
        
        # Extract text from next_start_page (up to next_subtitle_index)
        if next_start_page is not None:
            next_page_text = extract_text_from_page(pdf_path, next_start_page)
            end_content = next_page_text[:next_subtitle_index] if next_subtitle_index is not None else next_page_text
            return start_content + " " + end_content
        else:
            return start_content  # No next page to consider
        
    else:  # Case 2: Extract content across a range of pages
        content = ""
        for page_num in range(start_page, next_start_page):
            content += extract_text_from_page(pdf_path, page_num)
        return content

# Path to your PDF file
pdf_path = "1.pdf"

# Initialize the 'content' column
df['content'] = None

# Iterate through the rows of the DataFrame
for index, row in df.iterrows():
    start_page = int(row['start_page']) - 1  # Convert start_page to 0-indexed
    subtitle_index = row['subtitle_index']
    
    # Determine the next_start_page and next_subtitle_index (if this is not the last row)
    if index < len(df) - 1:
        next_start_page = int(df.iloc[index + 1]['start_page']) - 1
        next_subtitle_index = df.iloc[index + 1]['subtitle_index']
    else:
        next_start_page = None
        next_subtitle_index = None
    
    # Extract content
    content = extract_content(pdf_path, start_page, subtitle_index, next_start_page, next_subtitle_index)
    
    # Store the content in the DataFrame
    df.at[index, 'content'] = content
```

```python
# Save the DataFrame to a CSV file
output_csv_path = "out_put.csv"
df.to_csv(output_csv_path, index=False)

print(f"Data saved to {output_csv_path}")
```
